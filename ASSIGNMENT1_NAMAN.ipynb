P1: a) Optimizing the Equation using Gradient Descent

```python
import numpy as np

# Define the function and its derivative
def f(x):
    return x**2 + 4*x + 4

def df(x):  # Derivative of f(x)
    return 2*x + 4

# Gradient Descent algorithm
def gradient_descent(learning_rate, n_iterations, initial_x):
    x = initial_x
    for i in range(n_iterations):
        gradient = df(x)
        x = x - learning_rate * gradient
        if i % 10 == 0:
            print(f"Iteration {i}: x = {x}, f(x) = {f(x)}")
    return x

# Parameters
learning_rate = 0.1
n_iterations = 100
initial_x = 10

# Finding the minima
minima = gradient_descent(learning_rate, n_iterations, initial_x)
print(f"Minima of the function is at x = {minima}")
```

P1: c) Random Data Arrays and Profit Calculation

```python
np.random.seed(0)  # Set seed for reproducibility
revenue = np.random.randint(10000, 50000, 15)
expenses = np.random.randint(9000, 27000, 15)

# Calculate Profit, Tax, and Margin
profit = revenue - expenses
tax = 0.3 * revenue
profit_after_tax = profit - tax
margin = (profit_after_tax / revenue) * 100

# Find good, bad, and worst months
mean_profit = np.mean(profit_after_tax)
good_months = np.where(profit_after_tax > mean_profit)[0]
bad_months = np.where(profit_after_tax < mean_profit)[0]
worst_month = np.argmin(profit_after_tax)

# Print results
print(f"Revenue: {revenue}")
print(f"Expenses: {expenses}")
print(f"Profit: {profit}")
print(f"Profit After Tax: {profit_after_tax}")
print(f"Margin: {margin}")
print(f"Good Months: {good_months}")
print(f"Bad Months: {bad_months}")
print(f"Worst Month: {worst_month}")
```

P2: a) Curve Fitting

```python
from scipy.optimize import curve_fit

# Example data: Processor Speed vs Price
processor_speed = np.array([1.4, 1.6, 1.8, 2.0, 2.2, 2.4, 2.6, 2.8, 3.0, 3.2])
price = np.array([100, 150, 200, 240, 290, 340, 390, 440, 500, 560])

# Define a model (linear in this case)
def model(x, a, b):
    return a * x + b

# Curve fitting
params, covariance = curve_fit(model, processor_speed, price)
print(f"Fitted parameters: {params}")

# Plot the fitted curve
plt.scatter(processor_speed, price, label='Data')
plt.plot(processor_speed, model(processor_speed, *params), label='Fitted Curve', color='red')
plt.xlabel('Processor Speed (GHz)')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

P2: b) Neural Network and Regression

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Generate random data
x = np.random.rand(200)
y = 2 * x + 3 + np.random.normal(0, 0.2, 200)

# Plot data points
plt.scatter(x, y)
plt.title('Random Data Points')
plt.show()

# Neural network
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=1))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))

# Compile model
model.compile(optimizer='sgd', loss='mse')

# Train the model
history = model.fit(x, y, epochs=100, verbose=0)

# Plot loss function vs epochs
plt.plot(history.history['loss'])
plt.title('Loss Function vs No of Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

P2: c) Compare Optimizers

```python
# Using SGD, Mini-batch SGD, and Adam
from tensorflow.keras.optimizers import SGD, Adam

optimizers = ['sgd', 'adam', 'RMSprop']

# Plot comparison for each optimizer
for opt in optimizers:
    model.compile(optimizer=opt, loss='mse')
    history = model.fit(x, y, epochs=50, verbose=0)
    plt.plot(history.history['loss'], label=opt)

plt.title('Loss Function Descent for Different Optimizers')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
